```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

url = "https://www.ebay.com/b/PC-Laptops-Netbooks/177/bn_317584"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/115.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

response = requests.get(url, headers=headers)
computers_list = []

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    products = soup.find_all("li", class_="brwrvr__item-card--list")

    for item in products[:10]:
        comp_name = item.find("a",{ 'class':"bsig__title__wrapper"}).text.strip()
        comp_price = item.find("span",{'class':"bsig__price--displayprice"}).text.strip()
        comp_price_before = item.find("span", {'class':"strikethrough"})
        comp_brand = item.find("span", {'class':"bsig__generic"}).text.strip()

      #  clipped
        computers_list.append({
            "اسم الكمبيوتر": comp_name,
            "السعر": comp_price,
            " السعر القديم": comp_price_before,
            "الشركة المصنعة": comp_brand ,
        })

        time.sleep(random.uniform(1, 2))

    dataset = pd.DataFrame(computers_list)
    print(dataset)
    # dataset
else:
    print("فشل تحميل الصفحة. رمز الاستجابة:", response.status_code)
```

```python
dataset.to_csv("pc_laptops.csv", index=False)
```
